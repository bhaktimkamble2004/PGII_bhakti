```python
import ee
```

```python
import geemap
```

```python
ee.Authenticate()
ee.Initialize(project='bhakti-ee')
```

```python
map = geemap.Map()
map
```

```python
TALUKA = ee.FeatureCollection('projects/bhakti-ee/assets/TALUKA_BOUNDARY')
map.addLayer(TALUKA, {}, "TALUKA_BOUNDARY")
map
```

```python
DEM = ee.Image('CGIAR/SRTM90_V4')
TALUKA_DEM = DEM.clip(TALUKA)
map.addLayer(TALUKA_DEM, {}, 'DEM')
map
```

```python
NDVI = ee.ImageCollection('LANDSAT/COMPOSITES/C02/T1_L2_8DAY_NDVI')
TALUKA_NDVI = NDVI.filterDate('2025-01-01', '2025-12-31').mosaic().clip(TALUKA)
map.addLayer(TALUKA_NDVI, {}, 'NDVI')
map
```

```python
Export_DEM = ee.batch.Export.image.toDrive(
    image=TALUKA_DEM,
    description="Export_DEM",
    folder="GEE_Exports",
    fileNamePrefix="DEM",
    region=TALUKA.geometry().bounds(),
    scale=30,
    crs="EPSG:4326",
    maxPixels=1e13,
    fileFormat="GeoTIFF"
)
Export_DEM.start()
```

```python
Export_DEM.status()
```

```python
Export_NDVI = ee.batch.Export.image.toDrive(
    image=TALUKA_NDVI,
    description="Export_NDVI",
    folder="GEE_Exports",
    fileNamePrefix="NDVI",
    region=TALUKA.geometry().bounds(),
    scale=30,
    crs="EPSG:4326",
    maxPixels=1e13,
    fileFormat="GeoTIFF"
)
Export_NDVI.start()
```

```python
Export_NDVI.status()
```
